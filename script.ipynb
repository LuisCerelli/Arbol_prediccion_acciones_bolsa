# Machine Learning Model to Predict Stock Market

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

path_file = '/Users/luisalbertocerelli/Desktop/00-Todo/01_Practicas_Data_Science/Data/Apple.xlsx'

df_apple = pd.read_excel(io=path_file, index_col='Date', parse_dates=True)


## Machine Learning Model
### Separate the data

#1. Target: which variable do you want to predict?
#2. Explanatory: which variables will you use to calculate the Prediction?

#colocamos la variable del objetivo

target = df_apple['Direction']

#colocamos las otras variables que seran las explicativas

explanatory = df_apple.drop(columns='Direction')

### Compute the model

T#he following Python code will compute the numbers of the mathematical equation that we will use to calculate if the ticker goes UP or DOWN.

#Creamos el modelo, y le colocamos un maximo de profundidad 3 para que se pueda leer mejor

model = DecisionTreeClassifier(max_depth=3)

#Ajustamos el modelo para que luego no dé errores

model.fit(X=explanatory, y=target)

#y ejecutamos el modelo.. 

model.predict(X=explanatory)

#creamos una tabla para comprar los datos de prediccion con los reales... 

df_pre = target.to_frame()

#y creo una nueva columna de Predicciones para ver y comparara las dos columnas juntas... 

df_pre['Prediction'] =  model.predict(X=explanatory)


### Visualize the model

#y aqui vemos el arbol, recordar que mas arriba habiamos colocado una profundidad de 3, lo que hace que sea visualice mejor, ademas, al colocar el ; al final de los )) hacemos que no se vea el texto de los calculos, para finalizar: filled le da color y fontsize el el tamano de la fuente

plot_tree(decision_tree=model,feature_names=explanatory.columns.to_list()); #filled=True, fontsize= 5)

### Calculate the Predictions

### Evaluate the model: compare Predictions with the reality
#comenzamos a evaluar si la prediccion coincide con la prediccion... para ello creamos una variable 'comp', de comparacion

comp = df_pre.Direction == df_pre.Prediction

# Entonces hacemos una suma que nos dice cuantos dias el modelo acertó con respecto a la realidad... 

comp.sum()

# y esta es la longitud de todos los dias que hubo... 

len(comp)

# y esta es la division que nos dira la probabilidad del modelo... 

comp.sum()/len(comp)

#y esta es una forma de hacerlo directamente, no de forma manual como se hizo mas arriba... 

model.score(X=explanatory, y=target)
```
